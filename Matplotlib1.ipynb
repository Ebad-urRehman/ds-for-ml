# Plots

Plots also known as charts are visual representation of data in form of colored(mostly) graphics.

# Types of plots
### Line Plot
### Bar Plot
### Scatter Plot
### Pie Plot
### Area Plot
### Histogram Plot

```python
import matplotlib.pyplot as plt
```

```python
# initializing data
x = [12, 13, 16, 27, 20]
y = [222, 113, 196, 827, 320]


# plotting the data
plt.plot(x,y)

# adding title
plt.title("Plot")

# adding x and y labels
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

```python
# initializing data
x = [12, 13, 16, 27, 20]
y = [222, 113, 196, 827, 320]


# plotting the data
plt.bar(x,y)

# adding title
plt.title("Plot")

# adding x and y labels
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

```python
# initializing data
x = [12, 13, 16, 27, 20]
y = [222, 113, 196, 827, 320]


# plotting the data
plt.scatter(x,y)

# adding title
plt.title("Plot")

# adding x and y labels
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

```python
# Data
ages = [22, 25, 27, 27, 28, 30, 31, 32, 33, 34, 35, 35, 35, 36, 37, 40, 41, 42, 42, 45]

# Create histogram
plt.hist(ages, bins=[20, 26, 31, 36, 41, 46], edgecolor='black')

# Add titles and labels
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Show plot
plt.show()
```

```python
# pie chart

x = [10, 20, 30, 40]

plt.pie(x, explode=(0., 0, 0.1, 0))

plt.title("Pie Chart")
```

```python
# 3d plot

figure = plt.figure()
ax = plt.axes(projection='3d') # created a 3d graph

x = [1, 2, 3]
y = [4, 5, 6]
z = [7, 8, 9]

ax.plot3D(x,y,z)
```
