```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# reading df using pandas
df = pd.read_csv('flights.csv', delimiter=",")
print(df)
```

```python
x = df['month']
y = df['passengers']

plt.plot(x,y)
plt.xlabel('Month')
plt.ylabel('Passengers')
plt.title('Monthly Passenger Traffic Trend')
plt.xticks(rotation=45)
```

```python
means = []
for month in df['month'].unique():
    means.append(df.loc[df['month']==month]['passengers'].mean())

print(means)
```

# Line graph


```python
x = df['month'].unique()
y = means

plt.plot(x,y)
plt.xlabel('Month')
plt.ylabel('Passengers')
plt.title('Monthly Passenger Average Traffic Trend')
plt.xticks(rotation=45)
```
