```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
```

```python
df = seaborn.load_dataset("tips")
print(df)

print(df.head()) # shows 1st 5 rows of the dataframe
print(df.tail()) # shows last 5 rows of the dataframe
```

# Swarm Plot

A swarm plot is a type of scatter plot that is used for representing categorical values.

```python
seaborn.swarmplot(x=df["tip"])
seaborn.set_style("darkgrid")
plt.show()
```

```python
# use pyplot to visualize tip observatiosn, and amounts by day of the week
seaborn.swarmplot(x='day', y='tip', data=df)
seaborn.set_style('whitegrid')
```

```python
gender_palette = ['#A833FF', '#FFAF33']
seaborn.swarmplot(x='day', y='tip', hue="sex", palette=gender_palette, data=df)
seaborn.set_style('whitegrid')
```

# Voilin Plot
statistical representation of numerical data
similiar to box plot with an addition of rotated kernel density plot on each side

```python
# single value plot i think
seaborn.violinplot(x=df['tip'], color='gold')
```

```python
# draw plot based on variable two axis
seaborn.violinplot(x="sex", y="total_bill", data=df)
```

# Facet Grids
Facet grid objects take dataframe as an input and the names of variables that will form row, column, or hue dimensions of the grid. The variables should be categorical and the data at each level of the variable will be used as a facet along that axis

```python
fg = seaborn.FacetGrid(df, col='time', row='sex') # make 4 graphs and named them
fg = fg.map(plt.hist,"tip",color="blue")
```

```python
fg = seaborn.FacetGrid(df, col='time', row='sex')
fg = fg.map(plt.scatter, "total_bill","tip",color="white", edgecolor="black")
```

```python
fg = seaborn.FacetGrid(df, col='time', row='sex')
fg = fg.map(plt.scatter, "total_bill","tip")
fg = fg.add_legend()
```
